# Data processing

```python
data.columns
```

```python
# Normalizing data using power transforms
from sklearn.preprocessing import PowerTransformer
cols = ['int.rate', 'installment', 'log.annual.inc',
       'dti', 'fico', 'days.with.cr.line', 'revol.bal', 'revol.util',
       'inq.last.6mths']
pt = PowerTransformer()
data[cols] = pt.fit_transform(data[cols])
data.head()
```

```python
data_purpose = pd.get_dummies(data['purpose'], drop_first = False)
```

```python
data = pd.concat([data, data_purpose], axis = 1 )
```

```python
data.head()
```

```python
data.drop('purpose', inplace = True, axis = 1)
```

```python
data.shape
```

```python
data.info()
```

# Preprocessing

```python
X = data.drop(['not.fully.paid'], axis = 1)
y = data['not.fully.paid']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,  X_test,y_train, y_test = train_test_split(X,y, train_size = 0.7 )
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# function to evaluate
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_auc_score, f1_score
from sklearn.model_selection import StratifiedKFold
def analyse(model, xtrain, ytrain, xtest, ytest):
    model.fit(xtrain, ytrain)
    sk = StratifiedKFold(5, shuffle=True, random_state=0)
    sf_scorer = cross_validate(model, xtrain, ytrain, 
                               scoring=['roc_auc', 'accuracy', 'precision', 'recall', 'f1'], cv=sk)
    accuracy = sf_scorer['test_accuracy'].mean()
    recall = sf_scorer['test_recall'].mean()
    precision = sf_scorer['test_precision'].mean()
    roc_auc = sf_scorer['test_roc_auc'].mean()
    f1score = sf_scorer['test_f1'].mean()
    print('Training Set Scores\n')
    print(f'Accuracy: {accuracy}\nRecall: {recall}\nPrecision: {precision}\nROC_AUC: {roc_auc} \nf1_score: {f1score}\n')
    ypred = model.predict(xtest)
    print('\nValidation Set Scores\n')
    print(f'Accuracy: {accuracy_score(ytest, ypred)}\nRecall: {recall_score(ytest, ypred)}\nPrecision: {precision_score(ytest, ypred)}\nROC_AUC: {roc_auc_score(ytest, ypred)} \nf1_score: {f1_score(ytest, ypred)}')
```

# Declaring Model with default Parameters

```python
#instantiate the model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(class_weight={0:1, 1:3})
```

```python
analyse(lr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB() # default hyperparameter settings
analyse(gnb, X_train, y_train, X_test, y_test)
```

# Hyperparameter Optimization

```python
params = {'var_smoothing': [1/(10**i) for i in range(0,10)]}
```

```python
from sklearn.model_selection import GridSearchCV
search = GridSearchCV(gnb,params, cv = 3, scoring = 'roc_auc', refit = True)
#start = time.time()
search.fit(X_train, y_train)
#end = time.time()
#fit_time = end - start
#print('Fit_time: ', fit_time)
```

```python
search.best_params_
```

```python
gnb = GaussianNB(**search.best_params_)
analyse(gnb, X_train, y_train, X_test, y_test)
```
